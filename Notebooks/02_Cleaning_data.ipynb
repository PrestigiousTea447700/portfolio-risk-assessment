```python
import pandas as pd
import os

assets = ['AAPL', 'MSFT', 'TLT', 'VNQ', 'SPY']
processed_folder = 'C:\\projects\\portfolio-risk-assessment\\Data\\processed\\'

dfs = []
for ticker in assets:
    df = pd.read_csv(f"{processed_folder}{ticker}_clean.csv", index_col=0, parse_dates=True)
    dfs.append(df)

# Combine side by side using date as index
combined = pd.concat(dfs, axis=1)
combined = combined.dropna()  # Keep only rows where all assets have data

# Save combined data for analysis
combined.to_csv(f"{processed_folder}portfolio_prices.csv")
print(combined.head())
```

```python
returns = combined.pct_change().dropna()
returns.to_csv(f"{processed_folder}portfolio_returns.csv")
print(returns.head())
```
